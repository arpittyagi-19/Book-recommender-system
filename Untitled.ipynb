### Popularity Based recommender system

```python
ratings_name = ratings.merge(books, on='ISBN')
```

```python
ratings_name
```

```python
ratings_name.shape
```

```python
num_rating_df = ratings_name.groupby('Book-Title').count()['Book-Rating' ].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'}, inplace = True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on = 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df['Image-URL-M'][0]
```

### Colaborative Filtering

```python
x = ratings_name.groupby('User-ID').count()['Book-Rating'] > 200
rating_users = x[x].index
```

```python
filtered_rating = ratings_name[ratings_name['User-ID'].isin(rating_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    suggestions=[]
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
    
```

```python
recommend('1984')
```
